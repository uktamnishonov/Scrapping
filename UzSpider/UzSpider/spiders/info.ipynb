```python
import pandas as pd
```

```python
with open('uz_articles.json') as f:
    uz = pd.read_json(f)
    
with open('ru_articles.json') as f:
    ru = pd.read_json(f)
    
with open('cyrl_articles.json') as f:
    cyrl = pd.read_json(f)
```

```python
# Converting the type of creation_date column to datetime
uz['creation_date'] = pd.to_datetime(uz['creation_date'])
ru['creation_date'] = pd.to_datetime(ru['creation_date'])
cyrl['creation_date'] = pd.to_datetime(cyrl['creation_date'])
```

```python
# Extracting the month from the creation_date column
uz['creation_month'] = uz['creation_date'].dt.month
ru['creation_month'] = ru['creation_date'].dt.month
cyrl['creation_month'] = cyrl['creation_date'].dt.month
```

```python
# Plotting the number of articles created in each month
uz['creation_month'].value_counts().sort_index().plot(kind='bar')
```

```python
ru['creation_month'].value_counts().sort_index().plot(kind='bar')
```

```python
cyrl['creation_month'].value_counts().sort_index().plot(kind='bar')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud().generate(' '.join(uz['title']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wordcloud = WordCloud().generate(' '.join(ru['title']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wordcloud = WordCloud().generate(' '.join(cyrl['title']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
